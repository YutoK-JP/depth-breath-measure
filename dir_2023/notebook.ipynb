```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def local_norm(array, width=50):
  if(len(array.shape)!=1 or array.size<width):
    return -1
  
  result = []
  for i in range(array.size-width):
    local_array = array[i:i+width]
    bot = local_array.min()
    top = local_array.max()
    normed = (local_array[-1]-bot)/(top-bot)
    result.append(normed)
  return np.array(result)
```

```python
data = np.load("data.npz")
```

```python
local_norm(data["y_chest"])
```

```python
width = 100

w_size = 7
v = np.ones(w_size) / w_size # ,1/7
out = np.convolve(data["y_stomach"], v, mode='same')

plt.subplot(311)
plt.plot(out[10:-10])
plt.subplot(312)
plt.plot(data["t"][width:], data["y_stomach"][width:])
plt.subplot(313)
plt.plot(data["t"][width:], local_norm(data["y_stomach"], width))
```
